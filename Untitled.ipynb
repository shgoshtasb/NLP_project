## Get Calibration plots

```python
import numpy as np
import torch
import pickle
import os

tasks = ['DailyDialog', 'HateOffensive', 'SarcasmGhosh', 'SentiTreeBank', 'ShortHumor', 'ShortRomance', 'StanfordPoliteness', 'TroFi', 'VUA'] #'ShortJokeKaggle', CrowdFlower
labelsym = [[
            "noemotion",
            "anger",
            "disgust",
            "fear",
            "happiness",
            "sadness",
            "surprise",
        ], ["hate", "offensive", "neither"], ["sarcastic", "literal"], [ "negative", "positive"],["humorous", "literal"],["romantic", "literal"], ["polite", "impolite"], ["metaphor", "non-metaphor"], ["metaphor", "non-metaphor"]]

import matplotlib.pyplot as plt
import matplotlib.cm as cm
%matplotlib inline  



import os 
os.chdir('code')
from calibration import torch_utils

for t, task in enumerate(tasks):
    fig, ax = plt.subplots(nrows=2, ncols=1, figsize=[5,10])
    output_dir = '../' + task.lower() + '_'
    task_name = task.lower()

    types = {'MLE':'', 'poscal':'poscal', 'platt':'platt', 'plattbin':'plattbin', 'plattbintop':'plattbintop'}
    
    datafile = '../xslue_data/{}/test.tsv'.format(task)
    with open(datafile, 'r') as f:
        sentences = f.read().split('\n')
        sentences = [sent[:-2] for sent in sentences]
        sentences = np.array(sentences)
            
    keys = list(types.keys())
    prop_cycle = plt.rcParams['axes.prop_cycle']
    colors = iter(prop_cycle.by_key()['color'])
    for k, key in enumerate(keys):
        color = next(colors)
        preds = os.path.join(output_dir + types[key], 'preds')
        with open(preds, 'rb') as f:
            logits, labels = pickle.load(f)

        indices = np.arange(len(labels))
        num_labels = logits.shape[1]
        
        bins = torch.linspace(0, 1., 11)
        p_hats = torch.softmax(torch.tensor(logits), dim=1)
        labels = torch.tensor(labels)
        
        confidence, preds = torch.max(p_hats, dim=1)
        p_binums = torch_utils.searchsorted(bins, confidence) - 1
        onehotbinindex = torch.eye(bins.shape[0] - 1)[p_binums]
        num_samples = onehotbinindex.sum(dim=0)
        onehot = torch.eye(num_labels)[labels]
        correct_preds = (preds == labels).to(torch.float32)
        accuracy = (onehotbinindex * correct_preds.reshape(-1, 1)).sum(dim=0)/(num_samples + 1e-20)
        bin_mean = (bins[1:] + bins[:-1])/2
        confidence_avg = ((onehotbinindex * confidence.reshape(-1, 1)).sum(dim=0)/(num_samples +1e-20)) * (num_samples>0).to(torch.float) + bin_mean * (1 - (num_samples>0).to(torch.float32))
        
        ax[0].plot(confidence_avg, accuracy, marker='o', markersize=5, c=color, linestyle=None)
        ax[0].plot(np.linspace(0., 1., 11), np.linspace(0., 1., 11), marker='o', markersize=5, c='gray', linestyle='--', label='_nolegend_')
        ax[1].plot(bins[:-1], num_samples, marker='o', markersize=5, c=color, linestyle=None)
        ax[0].set_ylabel('Accuracy', fontsize=18)
        ax[1].set_ylabel('# Samples', fontsize=18)
        ax[1].set_yscale('log')
        ax[1].set_xlabel('Confidence average', fontsize=18)

        ax[0].set_title(task, fontsize=18)
    ax[1].legend(keys)
    plt.subplots_adjust(wspace=None, hspace=None)
    plt.savefig('calibrationplots_{}.png'.format(task), format='png', dpi=600, transparent=True, bbox_inches='tight', pad_inches = 0)
```
